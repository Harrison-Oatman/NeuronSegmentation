```python
import matplotlib.pyplot as plt

from runner import *

plotpath = parent_directory + "\\plots\\"
plotting.set_dir(plotpath)
```

```python
def get_within_bounds(rna):

    rec_rna = rna[rna['global_y'] < bbox_ymax]
    rec_rna = rec_rna[rec_rna['global_y'] > bbox_ymin]
    rec_rna = rec_rna[rec_rna['global_x'] < bbox_xmax]
    rec_rna = rec_rna[rec_rna['global_x'] > bbox_xmin].copy()

    rec_rna['global_y'] -= bbox_ymin
    rec_rna['global_x'] -= bbox_xmin

    return rec_rna
```

```python
pre_segmentation_image = np.array((thresholdedProcessImage > 0) + 2 * (somaImage > 0), dtype=float)
```

```python
def plot_barcode(rna, id):
    barcode_pts = rna[rna['barcode_id'] == id]
    fig, ax = plotting.colorplot(plotting.panel(pre_segmentation_image, bbox, 1), two_map, interpolation="None",
                   name="temp", show=False)
    ax.scatter(barcode_pts['global_x'], barcode_pts['global_y'], c='k', s=1)

    plt.savefig(plotpath + f"plotted_barcode_{id}.png")
    plt.show()

def plot_global(rna, id):
    barcode_pts = rna[rna['barcode_id'] == id]
    fig, ax = plotting.colorplot(pre_segmentation_image, two_blue, interpolation="None",
                   name="temp", show=False)
    ax.scatter(barcode_pts['global_x'], barcode_pts['global_y'], c=two_orange[1], s=1)

    plt.savefig(plotpath + f"global_plotted_barcode_{id}.png")
    plt.show()

def plot_two_global(rna, idone, idtwo):
    barcode_pts_one = rna[rna['barcode_id'] == idone]
    barcode_pts_two = rna[rna['barcode_id'] == idtwo]
    fig, ax = plotting.colorplot(pre_segmentation_image, two_blue, interpolation="None",
                   name="temp", show=False)
    ax.scatter(barcode_pts_one['global_x'], barcode_pts_one['global_y'], c=two_orange[1], s=1.3)
    ax.scatter(barcode_pts_two['global_x'], barcode_pts_two['global_y'], c='g', s=1.3)

    plt.savefig(plotpath + f"global_plotted_barcode_{idone}_{idtwo}.png")
    plt.show()

plot_global(RNA, 1170)
# plot_two_global(RNA, 541, 1207) #gad2, slc17a7
# plot_two_global(RNA, 702, 745) #gfap, pdgfra
```
